```python
from si.io.csv_file import read_csv
from si.model_selection.split import train_test_split
from si.models.ridge_regression_least_squares import RidgeRegressionLeastSquares
```

```python
data= read_csv(r'C:\Users\anali\PycharmProjects\si\datasets\cpu\cpu.csv',sep=",",label=True, features=True)
```

```python
#Standar data
from sklearn.preprocessing import StandardScaler
data.X = StandardScaler().fit_transform(data.X)
```

```python
# Split the data into training and testing sets
train_dataset, test_dataset = train_test_split(data, test_size=0.2, random_state=42)
```

```python
test_dataset.y
```

```python
test_dataset.X
```

```python
# Create a Ridge Regression model and fit it
model = RidgeRegressionLeastSquares(l2_penalty=0.1, scale=True)
model.fit(train_dataset.X, train_dataset.y)
```

```python
# Make predictions and calculate the MSE
y_pred = model.predict(test_dataset.X)
mse = model.score(test_dataset.X, test_dataset.y)
```
